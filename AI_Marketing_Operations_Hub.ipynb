```python
import pandas as pd

import numpy as np

df = pd.read_csv(

    "marketing_campaign.csv",

    sep=";"

)

df.head()
```

```python
df.columns.tolist()
```

```python
df = df.drop_duplicates()

df = df.dropna()

df.shape
```

```python
df["Total_Spend"] = (

    df["MntWines"] +

    df["MntFruits"] +

    df["MntMeatProducts"] +

    df["MntFishProducts"] +

    df["MntSweetProducts"] +

    df["MntGoldProds"]

)

df["Total_Purchases"] = (

    df["NumWebPurchases"] +

    df["NumCatalogPurchases"] +

    df["NumStorePurchases"]

)

df["Campaign_Engagement"] = (

    df["AcceptedCmp1"] +

    df["AcceptedCmp2"] +

    df["AcceptedCmp3"] +

    df["AcceptedCmp4"] +

    df["AcceptedCmp5"] +

    df["Response"]

)
```

```python
def segment_customer(row):

    if row["Total_Spend"] > df["Total_Spend"].quantile(0.75):

        return "High Value"

    elif row["Total_Spend"] > df["Total_Spend"].median():

        return "Medium Value"

    else:

        return "Low Value"

df["Customer_Segment"] = df.apply(

    segment_customer,

    axis=1

)
```

```python
summary = (

df.groupby("Customer_Segment")

.agg(

Customers=("ID","count"),

Average_Spend=("Total_Spend","mean"),

Average_Purchases=("Total_Purchases","mean"),

Campaign_Response=("Response","sum")

)

.reset_index()

)

summary
```

```python
summary = (

df.groupby("Customer_Segment")

.agg(

Customers=("ID","count"),

Average_Spend=("Total_Spend","mean"),

Average_Purchases=("Total_Purchases","mean"),

Campaign_Response=("Response","sum")

)

.reset_index()

)

summary
```
